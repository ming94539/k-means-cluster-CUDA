```python
# Scaling N
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#16, 128, 1024, 2048, 8192
#Matrix vector multiplication time
python = [0.28368592262268066, 0.40253400802612305, 5.481091022491455, 231.18355202674866, 25977.751570224762]
# ^ python run times with different iterations, n, and k
s_l = [3.09816e-05,0.000113808,0.00205578, 0.0197591,0.199265,1.9596, 19.8035 ] #Serial C runtimes
p_l = [ 0.0065515, 0.00597323,0.00625051,0.0115333,0.057346,0.404053, 3.72182] #Parallel Cuda 
sm_l = [0.00456951,0.00458606,0.00512805,0.0102402,0.050414,0.51033,3.80274] #Shared Memory
#Serial and Cuda have the same n, i, and k. 
s = pd.Series(s_l)
p = pd.Series(p_l)
sm = pd.Series(sm_l)
x = pd.concat([s,p,sm])
x = pd.DataFrame([s,p,sm], index = ['CPU','GPU','Shared Memory'])
x.columns= [10, 50, 1000, 10000, 100000, 1000000, 10000000]
x.head()
plt.plot([10, 50, 1000, 10000, 100000, 1000000, 10000000],s_l, label='C (CPU)')
plt.plot([10, 50, 1000, 10000, 100000, 1000000, 10000000], p_l, label='Cuda (GPU)',linewidth=3)
plt.plot([10, 50, 1000, 10000, 100000, 1000000, 10000000], sm_l, label='Shared Memory (GPU)')
plt.xlabel('N data points')
plt.ylabel('Seconds')
plt.legend()
plt.show()
plt.savefig('total.png')
```

```python
# Scaling K
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#16, 128, 1024, 2048, 8192
s_l = [0.195248, 1.3681,12.33, 24.39,245.4] #Serial C runtimes
p_l = [0.0490219, 0.0396047,0.0372351,0.0598304,0.477184] #Parallel Cuda 
sm_l = [0.0503132,0.0385979,0.0268585,0.045248,0.000518762] #Shared Memory
#Serial and Cuda have the same n, i, and k. 
s = pd.Series(s_l)
p = pd.Series(p_l)
sm = pd.Series(sm_l)
x = pd.concat([s,p,sm])
x = pd.DataFrame([s_l,p,sm], index = ['CPU','Global Mem GPU','Shared Memory'])
x.columns= [5, 50, 500, 1000,10000]
x.head()
plt.plot([5, 50, 500, 1000,10000],s_l, label='C (CPU)')
plt.plot([5, 50, 500, 1000,10000], p_l, label='Global memory (GPU)',linewidth=3)
plt.plot([5, 50, 500, 1000,10000], sm_l, label='Shared Memory (GPU)',linewidth=1)

plt.xlabel('K Clusters/Centroids')
plt.ylabel('Seconds')
plt.legend()
plt.show()
plt.savefig('total.png')
```
